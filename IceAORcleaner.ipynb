```python
df_filtered = AORdata_df[AORdata_df["office_name"].str.contains("Field Office", na=False, case=False)]
```

```python
df_filtered["area"] = df_filtered["area"].str.split(", ")
```

```python
print(df_filtered)
```

```python
us_states = [
    # Multi-word states first to prevent partial matches
    "New Hampshire", "New Jersey", "New Mexico", "New York",
    "North Carolina", "North Dakota",
    "South Carolina", "South Dakota",
    "West Virginia",
    "Rhode Island",
    # Single-word states after
    "Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado",
    "Connecticut", "Delaware", "Florida", "Georgia", "Hawaii", "Idaho",
    "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky", "Louisiana",
    "Maine", "Maryland", "Massachusetts", "Michigan", "Minnesota",
    "Mississippi", "Missouri", "Montana", "Nebraska", "Nevada",
    "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Tennessee", "Texas",
    "Utah", "Vermont", "Virginia", "Washington", "Wisconsin", "Wyoming"
]

pattern = r'\b(' + '|'.join(us_states) + r')\b'

df_filtered["state"] = df_filtered["area"].str.findall(pattern)
```

```python
print(df_filtered)
```

```python
df_final = df_filtered[["office_name", "office_name_short", "area"]]
```
